```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset=sns.load_dataset("tips")
dataset.head()
```

```python
sns.catplot(x="tip", y="size",data=dataset)
plt.show()
```

```python
limited_dataset=sns.load_dataset("tips").head(20)
limited_dataset
```

```python
sns.catplot(x="tip", y="size",data=limited_dataset,hue="sex",height=7,aspect=2)
plt.show()
```

```python
sns.catplot(x="tip", y="size",data=limited_dataset,hue="sex",kind="bar",height=7,aspect=2)
plt.show()
```

```python
sns.catplot(x="day", y="size",data=limited_dataset,hue="sex",kind="bar",height=7,aspect=2)
plt.show()
```

```python
sns.catplot(x="day", y="size",data=limited_dataset,hue="sex",kind="violin",height=7,aspect=2)
plt.show()
```

```python
sns.catplot(x="day", y="size",data=limited_dataset,hue="sex",kind="boxen",height=7,aspect=2)
plt.show()
```
