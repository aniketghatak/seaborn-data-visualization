# Line Plot Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
a=[1,2,3,4,5,6,7,8]
b=[2,3,4,5,6,7,8,10]
```

```python
#matplotlib chart
plt.plot(a,b)
plt.show()
```

```python
import pandas as pd
```

```python
var_a=[11,22,33,44,55,66,77,88]
var_b=[1,2,3,4,5,6,7,8]
```

```python
dataf=pd.DataFrame({"var_a":var_a,"var_b":var_b})
sns.lineplot(x="var_a",y="var_b",data=dataf)
```

```python
data=sns.load_dataset("penguins")
data.head()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=data,hue="sex",style="sex",markers=["o",">"],dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="body_mass_g",data=data,hue="sex",style="sex")
plt.show()
```

```python
# working with limited amount of data
data_limited=sns.load_dataset("penguins").head(50)
data_limited
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=data_limited,hue="sex",style="sex",markers=["o",">"],dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="body_mass_g",data=data_limited,hue="sex",style="sex",markers=["o",">"],dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="body_mass_g",data=data_limited,hue="sex",style="sex",markers=["o",">"],dashes=False)
plt.grid()
plt.show()
```
