# Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.violinplot(y=data["total_bill"])
plt.show()
```

```python
sns.violinplot(y=data["tip"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,hue="time",linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=data,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=data,order=["Female","Male"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,hue="sex",split=True,linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="smoker",y="total_bill",data=data,order=["Yes","No"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=data,hue="smoker",split=True,linewidth=2,palette="Dark2")
plt.show()
```
