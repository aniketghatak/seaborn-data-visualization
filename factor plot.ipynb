# Factor Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.factorplot(x="size",y="tip",data=data)
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=data,hue="sex")
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=data,hue="sex",kind="bar")
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=data,hue="sex",kind="box")
plt.show()
```
