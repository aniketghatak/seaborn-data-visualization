```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset("penguins")
data.head()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="flipper_length_mm",data=data,hue="sex")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="species",y="bill_length_mm",data=data,hue="sex")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="species",y="bill_depth_mm",data=data,hue="sex")
plt.show()
```
