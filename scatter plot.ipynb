# Scatter Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("penguins")
data.head()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex")
plt.show()
```

```python
# limit the dataset
limited_data=sns.load_dataset("penguins").head(25)
limited_data
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=limited_data,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=limited_data,hue="sex",style="sex",size="sex",sizes=(100,40),
                palette="ocean_r",alpha=1)
plt.show()
```
