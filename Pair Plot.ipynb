```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data,hue="sex")
plt.show()
```

```python
sns.pairplot(data,hue="sex",hue_order=["Male","Female"],markers=["*","<"])
plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(data,hue="time")
plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="time")
plt.show()
```

```python
sns.pairplot(data,hue="smoker")
plt.show()
```

```python
sns.pairplot(data,vars=["total_bill","tip"],hue="smoker")
plt.show()
```
