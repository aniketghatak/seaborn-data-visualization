```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",palette="crest",edgecolor="r",jitter=5,size=10)
plt.show()
```

```python
sns.stripplot(x="sex",y="total_bill",data=data,hue="sex",palette="crest",edgecolor="r",jitter=5,size=10)
plt.show()
```

```python
sns.stripplot(x="smoker",y="total_bill",data=data,hue="sex",palette="crest",edgecolor="r",jitter=5,size=10)
plt.show()
```

```python
sns.stripplot(x="time",y="total_bill",data=data,hue="sex",palette="crest",edgecolor="r",jitter=5,size=10)
plt.show()
```
